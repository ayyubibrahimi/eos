```python
import pandas as pd
from lib import standardize_item_numbers, standardize_col_names
```

```python
def read():
    dfa = (pd.read_csv("../data/calls_for_service/calls_for_service_2022_9_14_2022.csv")
           .pipe(standardize_col_names)
           .rename(columns={"nopd_item": "item_number"})
           .pipe(standardize_item_numbers, ["item_number"]))
    
    dfb = pd.read_csv("../data/real_time_crime_center/rtcc.csv").pipe(standardize_item_numbers, ["item_number"])
    df = pd.merge(dfa, dfb, on="item_number")
    return df
```
