```python
import pandas as pd
from lib import standardize_item_numbers
import numpy as np
```

```python
def clean():
    dfa = pd.read_csv("../data/police_reports/electronic_police_reports_2018_2022_names.csv")
    dfa = dfa.pipe(standardize_item_numbers, ["item_number"])

    dfb = pd.read_csv("../data/real_time_crime_center/rtcc.csv")
    dfb["rtcc_requested"] = "1"
    dfb = dfb.pipe(standardize_item_numbers, ["item_number"])

    df = pd.merge(dfa, dfb, on="item_number")
    return df
```

```python
df = clean()
```

```python
df
```

```python
def get_top_200(df, n=200):
    counts = df.groupby("uid").rtcc_requested.count()/len(df)
    pct = n/len(counts)
    return sorted(counts)[-n:], pct

get_20, pct = get_top_200(df)
get_20 = pd.DataFrame(get_20)
print(get_20.sum())
print(pct)
```
